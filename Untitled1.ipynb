```
#Random Tree classifier creates many decision trees and pick the values that are returned by most decision trees

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```
digits = load_digits()
dir(digits)
```

```
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i])
```

```
df = pd.DataFrame(digits.data)
df.head()
```

```
df['target'] = digits.target
```

```
df[0:12]
```

```
X = df.drop('target',axis='columns')
y = df.target
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```
model = RandomForestClassifier(n_estimators=20)  #20 random trees
model.fit(X_train, y_train)
```

```
model.score(X_test, y_test)
```

```
y_predicted = model.predict(X_test)
```

```
#to check how well the model performs
cm = confusion_matrix(y_test, y_predicted)  #actual target,predited target
cm
```

```
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')      #35 zeros n actual target,all 35 predicted correctly,  
```
